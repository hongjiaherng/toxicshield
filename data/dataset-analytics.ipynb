```python
import kagglehub
from kagglehub import KaggleDatasetAdapter
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# EDA (Toxic Tweets Dataset)

```python
# Load the specific CSV file directly into a DataFrame
df01 = kagglehub.dataset_load(
    KaggleDatasetAdapter.PANDAS,
    "ashwiniyer176/toxic-tweets-dataset",
    "FinalBalancedDataset.csv"
)
```

```python
print("--- Dataset Information ---")
print(df01.info())
print("\n--- Summary Statistics ---")
print(df01.describe())
```

## Visualisation

```python
# dataframe of Toxic Tweets Dataset
mock_df = df01

plt.figure(figsize=(8, 6))
sns.countplot(x='Toxicity', data=mock_df, palette='magma')
plt.title('Frequency of Toxic (1) vs. Non-Toxic (0) Tweets Comments')
plt.xlabel('Toxicity')
plt.ylabel('Count')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.savefig('toxicity_histogram.png')
print(mock_df.describe(include='all'))
```
